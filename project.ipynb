```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("D:/Folder/Data Analysis/datasets/Global YouTube Statistics.csv",  encoding='ISO-8859-1')
print(data)
top_subscribers = data[['Youtuber', 'subscribers']].sort_values(by='subscribers', ascending=False).head(10) #to sort top youtubers by subscribers in descending order 
sns.barplot(data = top_subscribers, x = 'subscribers', y = 'Youtuber')
plt.title('Top 10 YouTubers by Subscribers')
plt.xlabel('Subscribers')
plt.ylabel('YouTubers')
plt.tight_layout()
plt.show()
```

```python
top_views = data[['Youtuber', 'video views']].sort_values(by = 'video views', ascending = False).head(10)
sns.barplot(data = top_views, x = 'video views', y = 'Youtuber', palette = 'spring')
plt.title('Top 10 YouTubers based on video views')
plt.xlabel('Views')
plt.ylabel('YouTubers')
plt.tight_layout()
plt.show()
```
