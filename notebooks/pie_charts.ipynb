```python
##  Top 10 YouTubers by Subscribers

# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("D:/folder/data analysis/datasets/Global YouTube Statistics.csv",  encoding='ISO-8859-1')

#count how many YouTubers are in each category
category_count = data["category"].value_counts()

#create a pie chart
plt.figure(figsize=(8,8))
colors = plt.cm.Set3.colors  # Built-in colormap
plt.pie(category_count, labels = category_count.index, autopct = '%1.1f%%', startangle = 140, colors = colors)
plt.title('Distribution of YouTubers by Category')
plt.tight_layout()
plt.savefig('YouTubers by Category.png')
plt.show()
```

```python
#count how many YouTubers are in each country
country_count = data["Country"].value_counts().head(10)

#create a pie chart
plt.figure(figsize=(8,8))
colors = plt.cm.Set3.colors  # Built-in colormap
plt.pie(country_count, labels = country_count.index, autopct = '%1.1f%%', startangle = 140, colors = colors)
plt.title('Top 10 YouTubers by Country')
plt.tight_layout()
plt.savefig('Top 10 YouTubers by Country.png')
plt.show()
```

```python
# Group by category and sum of subscribers
categories = data.groupby("category")['subscribers'].sum().sort_values(ascending=False)

# to display only top 8 categories to avoid clutter
top_categories = categories.head(8)
others = categories[8:].sum()
top_categories['Others'] = others

# plot a pie chart
plt.figure(figsize=(8,8))
colors = plt.cm.Set3.colors  # Built-in colormap
plt.pie(top_categories, labels = top_categories.index, autopct = '%1.1f%%', startangle = 140, colors = colors)
plt.title('Top 8 YouTubers by Total Subscribers', fontsize = 14)
plt.tight_layout()
plt.savefig('Top 8 YouTubers by total subscribers.png')
plt.show()
```
