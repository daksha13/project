```python
## 🔝 Top 10 YouTubers by Subscribers

# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("D:/folder/data analysis/datasets/Global YouTube Statistics.csv",  encoding='ISO-8859-1')

# Sort the dataset to find top 10 YouTubers by subscriber count
top_subscribers = data[['Youtuber', 'subscribers']].sort_values(by='subscribers', ascending=False).head(10)

# Set the visual style
sns.set(style="whitegrid")

# Create a barplot
plt.figure(figsize=(10, 6))
sns.barplot(data=top_subscribers, x='subscribers', y='Youtuber', palette='viridis')

# Add plot titles and label
plt.title('Top 10 YouTubers by Subscribers', fontsize=14)
plt.xlabel('Subscribers (in millions)')
plt.ylabel('YouTuber')
plt.tight_layout()
plt.savefig("Top 10 YouTubers by Subscribers.png")
plt.show()
```

```python
# Sort the dataset to find top 10 YouTubers by video views
top_views = data[['Youtuber', 'video views']].sort_values(by = 'video views', ascending = False).head(10)

# Set the visual style
sns.set(style="whitegrid")

# Create a barplot
plt.figure(figsize=(10, 6))
sns.barplot(data = top_views, x = 'video views', y = 'Youtuber', palette = 'viridis')

# Add plot titles and label
plt.title('Top 10 YouTubers based on video views', fontsize=14)
plt.xlabel('Views (in  millions)')
plt.ylabel('YouTuber')
plt.tight_layout()
plt.savefig("Top 10 YouTubers by views.png")
plt.show()
```

```python
# Sort the dataset to find top 10 YouTubers by video views
no_of_videos = data[['Youtuber', 'uploads']].sort_values(by = 'uploads', ascending = False).head(10)

# Set the visual style
sns.set(style="whitegrid")

# Create a barplot
plt.figure(figsize=(10, 6))
sns.barplot(data = no_of_videos, x = 'uploads', y = 'Youtuber', palette = 'viridis')

# Add plot titles and label
plt.title('Top 10 YouTubers based on number of videos', fontsize=14)
plt.xlabel('Videos')
plt.ylabel('YouTuber')
plt.tight_layout()
plt.savefig("Top 10 YouTubers by number of videos.png")
plt.show()
```
